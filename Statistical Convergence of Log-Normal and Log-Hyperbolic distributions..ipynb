```python
import sys
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from extreme_value_sampling import *

# sample from log normal distribution
def log_normal(N,mu,sigma):
    log_normal_samples = np.random.lognormal(mu, sigma, N) 
    return log_normal_samples
```

<b> Emperical Mean Convergence for 2 cases of Log-Normal and Log-Hyperbolic: </b>

<i> Log-Normal Cases: </i>

```python
#Log normal Cases

#Case 1
N= [10, 20, 30, 50, 100, 200, 500, 1000]
B=200
for i in N:
    Mean_Log_Case1 = []
    Var_Log_Case1 = []
    quant_lower=[]
    quant_upper=[]
    for _ in range(B):
        log_normal_samples = log_normal(i,1,1)
        quant_lower.append(np.quantile(np.log(log_normal_samples), 0.05))
        quant_upper.append(np.quantile(np.log(log_normal_samples),0.95))
        Mean_Log_Case1.append(np.mean(np.log(log_normal_samples)))
        Var_Log_Case1.append(np.var(np.log(log_normal_samples)))
    plt.hist(Mean_Log_Case1, label='mean')
    plt.hist(quant_upper, label='95% quantile')
    plt.hist(quant_lower, label='5% quantile')
    plt.title('N='+str(i)+', Lognormal Case1')
    plt.legend()
    plt.show()
```

<i> Log-normal distribution: </i>
We see from the 2 cases of log-normal distributions that the mean value remains relatively constant (closer to theoreticl mean) as N increases. The standard deviation however, seems to get lower with increasing N, as evident by the decreasing width of the bell-curve. The distribution of 'B' number of random values in empirical mean and variance seem to follow Gaussian distribution. Overall, there seems to be a positive correlation between N and the convergence of empirical mean and variance.

<i> Log-hyperbolic distribution: </i>
We see from the 2 cases of log-hyperbolic distributions that the random empirical mean and variance do not follow Gaussian distribution as much as the Log-normal. They seem to be more positively skewed. The empiral mean value seems to be further away from the theoretical mean as N increases, and fluctuates more than the Log-normal distributions. However, like log-normal, the standard deviation seems to get lower with increasing N, as evident by the decreasing curve width. Overall, with increasing N there seems to be a positive correlation of precision but not accuracy.
